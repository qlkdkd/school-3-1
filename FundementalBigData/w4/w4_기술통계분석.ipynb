## 5-1. 평균, 분산, 표준편자
### 1. 평균
* 데이터의 중심 성향을, 분산과 표준편차는 데이터의 퍼짐 정도를 나타낼 수 있는 대표적인 기초 통계량
* 평균이라는 의미에는 중심(center)의 의미를 내포하고 있기 때문에 자료의 대표성을 나타내기에 적절
* 평균이 데이터의 중심의 의미를 갖고 있지만, 이러한 평균의 의미를 보다 명확하게 나타내기 위해서는 표준편차와 분산과 같은 통계량을 함께 고려하는 것이 필요



### 2. 분산, 표준편차
* 분산과 표준편차는 모두 자료의 퍼짐 정도를 나타내는 통계량으로서 산술평균이 갖지 못하는 자료의 산포도를 제공해줄 수 있는 지표
* 표준편차는 분산의 값을 제곱근한 것



### 실습
A쇼핑은 고객에게 쿠폰으로 증정된 할인권의 사용 횟수에 대해 평균과 분ㅅ간 및 표준편차를 성별에 따라 구해보고자 한다. 할인권의 사용 횟수 평균과 분산에 따라 고객에게 제공하는 쿠폰 제공 횟수를 조절할 수 있고, 만약 성별에 따라 할인권 사용 횟수에 대한 분포가 다르다면 성별에 따른 차별화된 할인쿠폰 제공 전략을 시행할 수 있기 때문이다.

* 편균, 분산, 표준편차 분석 과정
할인권 사용 횟수와 성별을 이용해 성별 할인권 사용 횟수의 평균, 분산, 표준편차를 구해본다.

```python
#모듈 및 데이터 탑재
import pandas as pd
df=pd.read_csv('Ashopping.csv', sep=',', encoding='CP949')
df_1=df[['할인권_사용 횟수', '성별']]

#성별 평균, 분산, 표준편차 구하기
print(df_1.groupby('성별').mean())#평균
print(df_1.groupby('성별').var())#분산
print(df_1.groupby('성별').std())#표준편차
```

* 결과 해석
* 할인권 사용 횟수는 약 16회, 표준편차는 약 8.3회라고 할 수 있다.
    * 이 패텅은 남성(0)과 여성(1)으로 구분하여 판단해도 큰 차이가 없다고 볼 수 있다.
    * 1년에 16회 이상의 평균적인 할인권 사용횟수를 보인다는 것은 우선 1년에 약 30회의 할인쿠폰을 제공하는 A쇼핑몰의 입장에서 할인쿠폰 전략이 우수한 성과를 보인다고 판단할 수 있다.

## 5-2. 왜도와 첨도
### 1. 왜도
* 자료들의 분포가 좌 또는 우로 치우쳐있는 정도를 나타내는 통계량
* 왜도 값이 0보다 큰 값을 가질 경우 전체적인 분포가 왼쪽으로 치우친 오른쪽 꼬리 비대칭 분포
* 왜도 값이 0보다 작은 값을 가질 경우에는 반대로 오른쪽으로 치우친 왼쪽 꼬리 비대칭 분포
* 왜도 값이 0일 경울에는 평균과 중앙값이 같으며 치우침이 없이 정확히 좌우 대칭인 분포



### 2. 첨도
* 분포의 뾰족한 정도를 나타내는 값으로 자료들이 얼마나 평균에 집중되어 있는지 확인할 수 있는 통계량
* 0을 기준으로 0보다 크면 정규분포보다 뾰족한 분포의 형태를 보이고, 0보다 낮을 경우 정규분포보다 완만한 구릉모양의 분포를 가짐



### 실습
A쇼핑몰의 서비스 만족도에 대한 왜도와 첨도를 살펴보자

```python
import pandas as pd
df=pd.read_csv('Ashopping.csv', sep=',', encoding='CP949')

print(df.서비스_만족도.skew())#왜도
print(df.서비스_만족도.kurt())#첨도

%matplotlib inline
df.서비스_만족도.hist(bins=7)
```

### 결과 해석
* A쇼핑몰의 서비스 만족도의 왜도 값은 -0.971로 도출되어 평균에 비해 약간 오른쪽으로 치우친 왼쪽 꼬리를 갖는 분포임을 확인할 수 있다.
* 또한 첨도 값은 1.148이 도출되어 정규분포보다 다소 뾰족한 형태의 분포임을 알 수 있다. 
* 이러한 결과는 히스토그램으로 표현된 분포 그래프를 통해서도 확인할 수 있다.
* A 쇼핑몰의 고객 서비스 만족도는 평균 값인 5.54에 비해 다소 높게 평가하는 고객들이 많다는 것을 알 수 있다.

## 3. 기타 기술통계량
### 1. 중앙값
* 자료를 크기 순서대로 늘어놓았을 때 가장 중앙에 위치한 값을 의미
* 자료의 대표성을 나타내는 보조적인 지표로 사용
* 자료의 수가 홀수일 경우 중앙값은 (자료의 수+1)/2의 값의 순번에 위치한 값
* 자료의 수가 짝수일 경우 중앙값은 자료의 수를 2로 나눈 값의 순번에 위치한 값과 (자료의 수+2)/2의 값의 순번에 위치한 값의 산술평균값



### 2. 최빈값, 사분위수
* 최빈값은 자료의 집단에서 가장 많이 관측되는 수치를 의미하며 중앙값과 마찬가지로 자료의 대표성을 나타내는 보조적인 지표
* 사분위수는 자료들의 크기 순서로 늘어놓았을 때 누적 백분율을 4등분한 각 점에 위치한 값을 의미



### 실습
앞 절의 사례에서와 같이 A쇼핑몰은 고객들에게 정기적으로 할인쿠폰을 발송하고 있다. 고객별로 매월 1장 내지는 2장 정도의 할인쿠폰을 발송하고 있는데, 고객들의 평균적인 할인쿠폰 사용빈도는 연 16회 정도이므로 현재 발송횟수가 적절하다고 생각해왔다. 그러나, 할인쿠폰 사용빈도의 평균만으로는 의사결정의 신뢰성이 떨어질 수 있기 때문에 중앙값, 최빈값 같은 기술 통계량으로 의사결정의 타당성을 검증해보고자 한다.

```python
import pandas as pd
import numpy as np
df=pd.read_csv('Ashopping.csv', sep=',', encoding='CP949')

#4분위수 구하기
print(np.percentile(df['할인권_사용 횟수'], 25))#1분위수
print(np.percentile(df['할인권_사용 횟수'], 50))#2분위수
print(np.percentile(df['할인권_사용 횟수'], 75))#3분위수
print(np.percentile(df['할인권_사용 횟수'], 100))#4분위수

#최빈값 출력
print('최빈값: ', df['할인권_사용 횟수'].value_counts().idxmax())

#기술통계량 출력
print("기술통계량: \n", df['할인권_사용 횟수'].describe())
```

### 결과 해석
* 고객들의 할인권 사용 횟수의 최빈값은 22회
* 사분위수의 1사분위는 9회, 2사분위는 17회(중앙값과 동일), 3사분위수는 22회
* 즉, 중간수준으로 할인쿠폰을 사용하는 고객은 연 17회 사용하며, 연간 22회를 사용하는 고객이 가장 많다는 것을 알 수 있음
* 전체적으로 상위고객으로 갈 수록 연간 할인권 사용 횟수는 많이 높아진다는 것을 알 수 있음음
