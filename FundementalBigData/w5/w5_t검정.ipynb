## 6-1. T검정 기초
### 1. T-검정의 개요
* 두 개 이하의 집단에서 수치형 변수의 평균의 일치성을 판단하는 기법으로 t-검정에 대한 설명에 앞서 평균비교 분석에 대해 살펴보고자 함

* 평균 비교분석
    * 평균을 비교하는 분석은 t검정, z검정, 분산분석 등이 있음
    * t-검정과 z-검정은 2개의 집단의 평균을 비교하는 분석이며, 분산분석은 3개 이상의 집단의 평균을 비교하는 분석
    * t-검정과 z-검정의 가장 큰 차이점은 모집단의 분산을 알고 있는지에 대한 여부에 따라 나뉨

## 6-2. 일표본 t검정
### 1. 일표본 t검정 개요
* 특정 수치로 알려진 모집단의 평균과 주어진 평균을 비교하기 위한 분석기법
* 우리 회사의 고객 평균 객단가는 10만원이 맞는가? 이 상품의 구매고객 평균 연령은 35세가 맞는가? 마케팅 영역에서는 지금까지 당연하게 받아들여졌던 특정 수치가 정말 맞는 것인지 검토해볼 수 있음.



### 2. 일표본 t검정 주요 이론
* 일표본 t검정의 검정통계량을 산출할 경우 모분산을 아는 경우(z-검정)과 모분산을 모르는 경우(t-검정)으로 나뉨
* 일표본 t-검정의 t값은 이론적 평균(모집단 평균)과 실제 평균(표본 평균)의 차이가 허용 가능한 오차 수준에(표준오차)에 비해 몇 배 더 큰지를 나타내는 값
* t값이 커질수록 모집단의 평균과 표본의 평균의 차이가 크다는 것임.



### 일표본 t검정의 분석지표
* 일표본 t-검정을 수행할 경우 결과 해석을 위해 확인해야 할 지표는 평균, t값, 그리고 Pr>|t|(유의확률) 등이 있음



### 실습
작년 A쇼핑의 VIP고객들의 평균 구매액은 약 700만원이었으며, 이에 A쇼핑은 올해 VIP 그룹 고객들의 평균 구매액 목표치를 10% 성장한 770만원으로 결정한 바 있다. 올해 VIP고객들의 평균 구매액 목표치인 770만원을 달성하였는지 판단하기 위해 일표본 t검정을 통해 검정해보고자 한다.

* 귀무가설: A쇼핑 VIP고객들의 총 매출액 평균은 770만원이다.
* 대립가설: A쇼핑 VIP고객들의 총 매출액 평균은 770만원이 아니다.

```python
import pandas as pd
from scipy import stats
df=pd.read_csv('Ashopping.csv', sep=',', encoding='CP949')

#총매출액 평균 및 일표본 t검정
print('총매출액 평균: ',df.총_매출액.mean())
print(stats.ttest_1samp(df['총_매출액'], 7700000))
```

### 결과 해석
* 일표본 t검정 결과 A쇼핑 vip고객들의 총 매출액 산술평균값은 7068690원이며
* t값은 -2.86, 유의확률은 0.0042로 도출되었다.
* 따라서, 귀무가설이 기각되어 A쇼핑 VIP고객들의 총 매출액 평균은 7700000이라고 할 수 없다.

* 또한 A쇼핑몰의 올해 VIP고객들의 매출액 평균이 약 706만원이었다.
* 이는 작년의 매출액 평균과 거의 차이가 없으며 목표치를 달성하지 못하였다는 결과이다.
* 따라서, A쇼핑몰이 목표치를 너무 과도하게 잡은 것이 아닌지 검토해보고, 경쟁사나 시장 평균 성장률과 비교했을 때 A쇼핑의 실적이 상대적으로 낮은 것이라면 VIP고객들의 매출액에 영향을 미치는 원인을 조사하여 적절한 매출신장 대책을 마련해야 할 것이다.

## 6-3. 독립표본 t검정

### 1. 독립표본 t검정의 개요
두 개의 독립된 그룹간의 특정 값의 평균을 비교하기 위한 분석기법
* 두 채널(on-line, off-line)의 고객 평균 객단가는 동일하다고 볼 수 있는가? 만약 두 채널의 평균 객단가가 유의한 차이를 보이고 있다면 각 쇼핑 채널에 대한 전략이 달라져야 할 것이다. 이러한 문제를 검토할 수 있도록 지원



### 2. 독립표본 t검정 주요 이론
* 독립된 두 표본의 분산이 같을 경우와 분산이 다를 경우 적용되는 공식이 다르다(원리는 같음)
* t값은 두 표본의 평균 차이 값과 두 모집단의 평균 차이 값간의 편차가 허용 가능한 오차 수준(표준 오차)에 비해 몇 배 더 큰지를 나타내는 값
* t값이 커질수록 두 집단의 평균의 차이가 크다는 것임.



### 3. 독립표본 t검정의 분석지표
* 독립표본 t검정을 수행할 경우 결과 해석을 위해 확인해야할 지표는 평균, 등분산성, f값, f값에 대한 유의 확률, 그리고 t값에 대한 유의확률이 있음



### 실습
A쇼핑의 마케팅 부서는 콜센터를 통해 접수된 클레임 고객에 대한 특별한 마케팅을 기획하고자 한다. 클레임 고객은 상대적으로 매장을 찾는 횟수가 적어져 이탈위험도가 높을 것이라고 예상되기 때문이다. 이를 위해 클레임 고객과 비클레임 고객간의 매장방문 횟수가 실제로 차이가 있는지를 먼저 검증해야 한다.

* 귀무가설: A쇼핑 클레임 고객들과 비클레임 고객들의 방문 횟수 차이는 없다.
* 연구가설: A쇼핑 클레임 고객들과 비클레임 고객들의 방문 쵯수 차이는 있다.

```python
import pandas as pd
from scipy import stats
import numpy as np
df=pd.read_csv("Ashopping.csv", sep=',', encoding='CP949')

#등분산검정
no_claim=df[df.클레임접수여부==0]
df2=np.array(no_claim.방문빈도)

claim=df[df.클레임접수여부==1]
df3=np.array(claim.방문빈도)
stats.bartlett(df2, df3)
```

```python
#1. 독립표본 t검정 및 방문빈도 평균
print(stats.ttest_ind(df2, df3, equal_var=False))
print('클레임 접수여부(0) 고객 평균방문빈도: ', no_claim.방문빈도.mean())
print('클레임 접수여부(1) 고객 평균방문빈도: ', claim.방문빈도.mean())
```

* 먼저 두 고객그룹간의 등분산성 검정 결과 F값은 13.626, 유의 확률은 0.00022로 도출되어 등분산 검정의 귀무가설이 기각됬음으로, 두 표본은 동일하다고 볼 수 없다.
* 따라서, 등분산이 가정되지 않은 t통계량을 확인하여 가설 검증을 수행해야 한다.

* 분석 결과, 클레임 접수 경험이 있는(claim=1)고객의 평균 방문 횟수는 24.73회, 클레임 접수 경험이 없는(claim=1)고객의 평균 방문 횟수는 28.18회이며
* 등분산 검정이 기각된 t값은 2.60, 유의 확률은 0.0096으로 도출되었다.
* 따라서, 귀무가설이 기각되어 클레임 접수 여부에 따라 방문 빈도는 차이가 있다고 할 수 있다.(유의수준 0.05수준)

* 사실 그 차이는 크지 않지만 통계적으로 유의한 차이가 있으므로 클레임 접수 고객에 대해 별도의 관리는 필요하다. 특히, 클레임 접수 고객이 VIP등급에 해당하는 고객이라면 클레임 접수 후 구매빈도나 방문 횟수가 점차 줄어드는지 확인하여 선제적인 대응을 통해 고객 유지 활동을 전개하는 것이 필요하다

## 6-4. 쌍체표본 t검정
### 1. 쌍체표본 t-검정 개요
* 동일집단의 실험 전과 실험 후 혹은 특정 시점과 그 이후 시점과 같이 쌍체 혹은 대응표본으로 이루어진 자료의 평균을 비교하기 위한 분석기법
* 동일한 고객집단의 실험 전과 실험 이후의 고객의 반응수치를 비교하고 실험 전과 후의 평균의 차이가 있는지 확인해볼 수 있음



### 2. 쌍체표본 t-검정 주요 이론
* 동일한 표본이 두 개의 서로 다른 상황(실험처리 여부 혹은 시간의 흐름)에서 특정 수치형 변수의 평균이 동일한지의 여부를 판단하는 분석방법
* 쌍체표본 t-검정에서 t값은 실험 전과 후 혹은 특정 시점과 그 이후 시점과 같이 쌍체로 이루어진 두 표본의 평균 차이 값과 이 두 쌍체 모집단의 평균 차이 값간의 편차가 허용 가능한 오차 수준(표준오차)에 비해 몇 배 더 큰지를 나타내는 값
* t값이 커질수록 두 쌍체 그룹의 평균의 차이가 크다는 것임



### 3. 쌍체표본 t-검정의 특징 및 결측값 처리 방법
* 독립표본 t-검정과 달리, 쌍체표본 t-검정은 분석대상의 표본이 반드시 대응되어야 함.
* 쌍체 표본 데이터는 시간상 전후의 개념이 있기 때문에 집단간의 독립성 가정은 필요하지 않으며, 두 모분산이 같다고 가정할 필요도 없음.
* 데이터가 완전한 쌍체를 이루지 않는다면 결측값이 존재하는 것이므로 아래와 같은 방법으로 결측값을 처리해야 함



### 4. 쌍체표본 t-검정의 분석지표
* 쌍체표본 t-검정을 수행할 경우 결과 해석을 위해 확인해야 할 지표는 평균, t값, 그리고 유의확률 등이 있음



### 실습
A쇼핑 마케팅 팀에서는 그 동안 단순 포인트 적립 및 결제시 차감 사용 방식으로 운영되어 온 멤버십 제도를 등급별 사은품 및 쿠폰 지급 기능이 추가된 전략적인 로열티 프로그램으로 개선하였다. 이를 통한 고객들의 로열티 프로그램 만족도의 변화가 통계적으로 유의한 지 알아보기 위해 고객들의 신규 멤버십 프로그램 도입 전과 도입 후의 고객 만족도에 대한 쌍체표본 t-검정을 수행하고자 한다.

* 귀무가설: A쇼핑 고객들의 멤버십 프로그램 도입 전 만족도와 도입 후 만족도의 차이는 없다.
* 대립가설: A쇼핑 고객들의 멤버십 프로그램 도입 전 만족도와 도입 후 만족도의 차이는 있다.

```python
import pandas as pd
from scipy import stats
df=pd.read_csv('Ashopping.csv', sep=',', encoding='CP949')

#쌍체표본 t-검정
stats.ttest_rel(df['멤버쉽_프로그램_가입후_만족도'], df['멤버쉽_프로그램_가입전_만족도'])
```

* t값: 29.560, 유의확률=0.0001이하(유의수준 0.05)
* 따라서 귀무가설이 기각되어 신규 프로그램의 가입 전 고객만족도와 가입 후 고객만족도는 차이가 있다고 할 수 있음

* 이를 통해 단순 포인트 중심의 멤버십 프로그램보다 고객 구매가 증대될수록 받는 사은품의 수준과 쿠폰 금액이 커지는 새로운 멤버십 프로그램을 고객들이 더 선호하여 전반적인 고객 만족도가 높아졌다는 점이 증명되었다.
* A쇼핑은 향후 기존에 적립되어 왔던 포인트 운영방식을 병행할 것인지, 아니면 고객 동의 하에 사은품과 쿠폰지급으로 누적 포인트를 소진시킬 것인지에 대한 정책적인 의사결정이 필요할 것이다.
